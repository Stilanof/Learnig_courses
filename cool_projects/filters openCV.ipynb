```python
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
uploaded=files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

img=cv2.imread("{name}".format(name=fn))

```

```python
cv2_imshow(img)
cv2.waitKey(4000)
```

**Using filters**

```python
# Change color pallet

img_c= cv2.cvtColor(img,4)
cv2_imshow(img_c)

# Blure

img_blr=cv2.GaussianBlur(img,(15,15),0)
cv2_imshow(img_blr)

# GaussianBlur
img_gb=cv2.GaussianBlur(img,(25,25),0)
cv2_imshow(img_gb)

#Bright
img_br=cv2.convertScaleAbs(img, beta=-40)
cv2_imshow(img_br)

#Canny
img_ca=cv2.Canny(img, 25, 25)
cv2_imshow(img_ca)

#Laplacian

img_ho=cv2.Laplacian(img,cv2.CV_8U)
cv2_imshow(img_ho)
```
